## **Data cleaning (Replace '--' to NaN)**

```python
data=data.replace('--',np.nan,regex=True)
```

```python
data.head(20)
```

## **Check Nulll values in the Dataset**

```python
data.isnull().sum()
```

```python
per_missing=data.isnull().sum()*100/len(data)
```

```python
per_missing
```

```python
sns.heatmap(data.isnull())
```

## **Drop null values**

```python
data.dropna(axis=0,inplace=True)
```

```python
sns.heatmap(data.isnull())
```

## **Data cleaning (Rank Column)**

```python
data.columns
```

```python
data['Rank']=data['Rank'].str[0:-2]
```

```python
data['Rank']=data['Rank'].str.replace(',','')
```

```python
data['Rank']=data['Rank'].astype('int')
```

```python
data['Rank'].dtypes
```

## **Data Cleaning [Video uploads &  Subscribers]**

```python
data.head()
```

```python
data.dtypes
```

```python
data['Video Uploads']=data['Video Uploads'].astype('int')
```

```python
data['Subscribers']=data['Subscribers'].astype(int)
```

```python
data.dtypes
```

## **Data Cleaning [Grade Column]**

```python
data.columns
```

```python
grade_mapping = {
    'A++': 5,
    'A+': 4,
    'A': 3,
    'A-': 2,
    'B+': 1
}

data['Grade'] = data['Grade'].astype(str).str.strip().map(grade_mapping).fillna(0).astype(int)
```

```python
data.tail()
```

## **Find AVG Views for Each Channel**

```python
data.columns
```

```python
data['Avg_views']=data['Video views']/data['Video Uploads']
```

```python
data.head()
```
